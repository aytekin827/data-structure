```python
# linked_list
# linked_list는 연결의 개념이 중요하다.
# create,add,del,print

class Node():
    def __init__(self,value):
        self.value = value
        self.next = None

class linked_list():
    def __init__(self,value):
        self.head = Node(value)

    def add_node(self,value):
        node = self.head
        while node:
            if node.next == None:
                node.next = Node(value)
                break
            else:
                node = node.next

    def del_node(self,value):
        node = self.head
        if self.head.value == value:
            print('head value deleted!')
            self.head = node.next
        else:
            while node:
                if node.next.value == value:
                    node.next = node.next.next
                    break
                else:
                    node = node.next
    
    def print_node(self):
        node = self.head
        while node:
            print(node.value)
            node = node.next

li = linked_list(5)
li.add_node(3)
li.add_node(4)
li.add_node(7)
li.add_node(9)
li.add_node(11)
li.add_node(33)
# li.head.next.value
li.print_node()
print()
li.del_node(4)
li.print_node()
li.head.next.value
```

```python
# stack
# pop,push,print
# first in last out LIFO
# linked list개념을 이용해서 stack data structure 구현하기

class linked_list():
    def __init__(self,value):
        self.value = value
        self.next = None

class stack():
    def __init__(self,value):
        self.top = linked_list(value)

    def push(self,value):
        new_node = linked_list(value)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        self.top = self.top.next

    def print_stack(self):
        node = self.top
        while node:
            print(node.value)
            node = node.next

st = stack(5)
st.push(3)
st.push(30)
st.push(300)
st.push(366)
st.push(36)
st.print_stack()
st.pop()
st.pop()
st.pop()
print()
st.print_stack()
```

```python
# queue
# FIFO first in first out 선입선출
# collection deque라는 모듈이 있지만 이걸 사용하지 않고 구현해보기
# linked_list, enqueue, dequeue

class linked_list():
    def __init__(self,value):
        self.value = value
        self.next = None

class queue():
    def __init__(self):
        self.front = None
        self.rear = None
    
    def enqueue(self,value):
        new_node = linked_list(value)

        if self.rear is None:
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.front is None:
            old_front = self.rear
            self.rear = None
        else:
            old_front = self.front
            self.front = self.front.next

    def print_queue(self):
        node = self.front
        while node:
            print(node.value)
            node = node.next

qu = queue()
qu.enqueue(4)
qu.enqueue(40)
qu.enqueue(400)
qu.enqueue(45)
qu.print_queue()

print()
qu.dequeue()
qu.print_queue()
```

```python
# binary search tree
# print할때는 왼쪽부터 찾아서 출력하도록 - 재귀함수 이용

class Node():
    def __init__(self,value):
        self.value = value
        self.right = None
        self.left = None

class binary_search_tree():
    def __init__(self,value):
        self.head = Node(value)

    def add_node(self,value):
        node = self.head
        while node:
            # 추가하려는 value가 작은경우
            if node.value > value:
                if node.left == None:
                    node.left = Node(value)
                    break
                else:
                    node = node.left
            # 추가하려는 value가 큰 경우
            else:
                if node.right == None:
                    node.right = Node(value)
                    break
                else:
                    node = node.right

    def search_node(self,value):
        node = self.head
        while node:
            # 추가하려는 value와 같은 경우
            if node.value ==  value:
                return True
                
            # 추가하려는 value가 큰 경우
            elif node.value < value:
                node = node.left
            else:
                node = node.right
        return False

def print_tree(node):
    # 재귀함수 이용
    
    if node:
        print(node.value)
        print_tree(node.left)
        print_tree(node.right)
        
bst = binary_search_tree(10)
bst.add_node(3)
bst.add_node(1)
bst.add_node(7)
bst.add_node(12)
bst.add_node(11)
bst.add_node(15)
bst.add_node(14)
bst.add_node(17)

# print(bst.head.right.right.right.value)
bst.search_node(100)
print_tree(bst.head)
```
